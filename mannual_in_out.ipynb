```python
%run baghchal_environment.ipynb
baghchal = BaghchalBoard()
    
while True:
    if baghchal.turn == 'G' and baghchal.goats_placed<20:  
        print(f"Goats remaining to be placed: {20-baghchal.goats_placed}") 
        print("Enter the position to place a goat .eg( 02) :")
        goat_position = input().strip()

        
        # Validate goat placement input
        if len(goat_position) == 2 and goat_position[0] in '01234' and goat_position[1] in '01234':
            
            start_row = int(goat_position[0])
            start_col = int(goat_position[1])
            end_row = int(goat_position[0])
            end_col = int(goat_position[1])
            
            if baghchal.is_valid_move(start_row,start_col,end_row,end_col) is  True :
                baghchal.make_move(start_row,start_col,end_row,end_col)
                baghchal.display_board()
            else:
                continue
                
                
         
                
        else:
            print("Invalid input. Try again.")
            continue

    else:
        print(f"{baghchal.turn}'s turn. Enter the move eg(00 01):")
        move_input = input().strip()
       
        # Validate move input
        if len(move_input) != 5:
            print("Invalid input.Try again.")
            continue
        
        start_position, end_position = move_input.split()
        start_row = int(start_position[0])
        start_col = int(start_position[1])
        end_row = int(end_position[0])
        end_col = int(end_position[1])
            
        
            
        if baghchal.is_valid_move(start_row,start_col,end_row,end_col) is True :
            baghchal.make_move(start_row,start_col,end_row,end_col)
            baghchal.display_board()
       
       # else:
        #    continue
    # Check win conditions
    result = baghchal.check_win_conditions()
    
    
    if result is None :
        continue
    
    else: 
        if result == 'Goat':
            print("Goats win!")
        elif result == 'Tiger':
            print("Tigers win!")
        break
    
```
